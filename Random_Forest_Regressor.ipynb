# **INTRODUCCIÓN A RANDOM TREE FOREST**

```python
import numpy as np
from google.colab import files
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
dt = files.upload()
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.info()
```

LINEAR REGRESSIÓN


```python
Y = df['sales']
X = df.drop('sales', axis = 1)
```

```python
model = LinearRegression()
```

```python
from sklearn import pipeline
pipeline = Pipeline([('model', LinearRegression())])
```

```python
pipeline.fit(X,Y)
```

```python
pipeline['model'].score(X,Y)
```

ARBOLES DE DECISIÓN

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
pipeline2 = Pipeline([('model', DecisionTreeRegressor())])
```

```python
pipeline2.fit(X,Y)
```

```python
pipeline2['model'].score(X,Y)
```

REPRETIR CON TRAIN TEST SPLIT 50/50


```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,train_size=.50,random_state=42)
```

ANALIZAMOS SCORES


```python
#ENTRENAMOS
pipeline.fit(X_train,Y_train)
```

R2 LINEAS REGRESSIÓN

```python
pipeline['model'].score(X_train,Y_train)
```

```python
pipeline['model'].score(X_test,Y_test)
```

```python
#ENTRENAMOS
pipeline2.fit(X_train,Y_train)
```

R2 DECISION TREE REGRESSOR

```python
pipeline2['model'].score(X_train,Y_train)
```

```python
pipeline2['model'].score(X_test,Y_test)
```

```python
print('=================================================')
print('-------------- Resumen de Scores ----------------')
print('==================================================')
print('\n1. Regresión Lineal:')
print(f'   - Score con todos los datos: {pipeline["model"].score(X,Y):.4f}')
print(f'   - Score en datos de entrenamiento (split): {pipeline["model"].score(X_train,Y_train):.4f}')
print(f'   - Score en datos de prueba (split): {pipeline["model"].score(X_test,Y_test):.4f}')

print('==================================================')

print('\n2. Árbol de Decisión (Decision Tree Regressor):')
print(f'   - Score con todos los datos: {pipeline2["model"].score(X,Y):.4f}')
print(f'   - Score en datos de entrenamiento (split): {pipeline2["model"].score(X_train,Y_train):.4f}')
print(f'   - Score en datos de prueba (split): {pipeline2["model"].score(X_test,Y_test):.10f}')

```
